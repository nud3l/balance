# Cost experiments

```python
import csv

agents = []
gas_curate = []
usd_curate = []
gas_update = []
usd_update = []


with open('./results/cost.3.csv') as csvfile:
    my_reader = csv.DictReader(csvfile, delimiter=',')
    for row in my_reader:
        if (row['gas_curate'] != '0'):
            agents.append(int(row['agents']))
            gas_curate.append(int(row['gas_curate']))
            usd_curate.append(float(row['usd_curate']))
            gas_update.append(int(row['gas_update']))
            usd_update.append(float(row['usd_update']))
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax1 = plt.subplots()
ax1.plot(agents, gas_curate, 'b', label="curate function")
ax1.plot(agents, gas_update, 'r', label="update function")
ax1.set_xlabel('Number of agents')
ax1.set_ylabel('Tx cost in gas')
ax1.tick_params('y')
plt.grid(True)

ax2 = ax1.twinx()
ax2.plot(agents, usd_curate, 'b.')
ax2.plot(agents, usd_update, 'r.')
ax2.set_ylabel('Tx cost in USD')
ax2.tick_params('y')

ax1.legend(loc=0)
plt.tight_layout()
plt.savefig("./figures/cost.png", dpi=600)
plt.show()
```
